```python
import numpy as np
import ngff_zarr as nz
```

```python
image = np.random.rand(512, 512)
```

```python
ngff_image = nz.to_ngff_image(
    data=image,
    dims=['y', 'x'],
    scale={'y': 0.5, 'x': 0.5},
    translation={'y': 10, 'x': 20},
    name='some_image',
    axes_units={'y': 'micrometer', 'x': 'micrometer'},
)
```

```python
ngff_multiscales = nz.to_multiscales(
    ngff_image,
    scale_factors=[
        {'y': 1.0, 'x': 1.0},
        {'y': 2.0, 'x': 2.0},
        {'y': 4.0, 'x': 4.0},
    ],
    method=nz.Methods.DASK_IMAGE_GAUSSIAN
)

ngff_multiscales
```

```python
nz.to_ngff_zarr(
    './test.ome.zarr',
    ngff_multiscales,
    version='0.6dev2'
)
```
